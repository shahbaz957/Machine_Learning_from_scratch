```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x,y = load_diabetes(return_X_y = True)
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , random_state = 2 , test_size=0.2)
```

```python
reg = LinearRegression()
reg.fit(x_train , y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
r2_score(y_test , y_pred)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
class stochastic_gd_regressor:
    def __init__(self , lr = 0.1 , epochs = 100):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs 
        self.lr = lr
    def fit(self , x_train , y_train):
        # init your coefficients
        self.intercept_ = 0
        self.coef_ = np.ones(x_train.shape[1])
        for i in range(self.epochs):
            for j in range(x_train.shape[0]):
                idx = np.random.randint(0 , x_train.shape[0])
                y_hat = np.dot(x_train[idx] , self.coef_) + self.intercept_
                
                ### for updating intercept_ value 
                loss_slope_intercept = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - self.lr*loss_slope_intercept

                ### for updating coefficient value 
                loss_slope_coef = -2 * np.dot((y_train[idx]-y_hat),x_train[idx])
                self.coef_ = self.coef_ - self.lr * loss_slope_coef
                
        print(self.intercept_ , self.coef_)
        
    def predict(self , x_test):
        return np.dot(x_test , self.coef_) + self.intercept_
```

```python
bg = stochastic_gd_regressor(lr = 0.01 , epochs = 30)
```

```python
bg.fit(x_train , y_train) # yes its working
```

```python
y_pred = bg.predict(x_test)
```

```python
r2_score(y_test , y_pred)
```
