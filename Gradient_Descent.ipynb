```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its gradient
def f(x):
    return x**2 + 3*x + 2

def gradient_f(x):
    return 2*x + 3  # derivative of x^2 + 3x + 2

# Gradient Descent Function
def gradient_descent(learning_rate=0.1, epochs=50, initial_x=5.0):
    x = initial_x
    x_history = [x]
    f_history = [f(x)]

    for i in range(epochs):
        grad = gradient_f(x)          # Compute gradient (slope)
        x = x - learning_rate * grad  # Update x
        x_history.append(x)
        f_history.append(f(x))
        print(f"Epoch {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

    return x, x_history, f_history

# Run it
final_x, x_steps, f_steps = gradient_descent()
```

```python
# Plot function and descent path
x_vals = np.linspace(-5, 5, 100)
y_vals = f(x_vals)

plt.figure(figsize=(10, 5))

# Plot the function
plt.plot(x_vals, y_vals, label='f(x) = x + 3x + 2')

# Plot the descent
plt.scatter(x_steps, f_steps, color='red', label='Descent Steps')
plt.plot(x_steps, f_steps, '--', color='red')

plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

# **Making a class of Gradient Descent**

```python
class gd_regressor():
    def __init__(self , lr , epochs):
        self.m = 100
        self.b = -120 # you can take any value which then gd manage itself
        self.lr = lr 
        self.epochs = epochs
    def fit(self , x , y):
        #calculate b using gd
        for i in range(self.epochs):
            loss_slope_b = -2 * np.sum(y - self.m * x.ravel() - self.b)
            loss_slope_m = -2 * np.sum((y - self.m * x.ravel() - self.b)*x.ravel())
            self.b = self.b - self.lr*loss_slope_b
            self.m = self.m - self.lr*loss_slope_m
        print(self.m , self.b)
    def predict(self , x):
        return self.m*x + self.b
            
            
        
```
